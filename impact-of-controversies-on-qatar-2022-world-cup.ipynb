## Background

An announcement made 11 years ago, "The winner to organize 2022 FIFA World Cup is Qatar", has been a source of controversies over the last decade. Outrages and proposed boycotts stemming from corruptions, human rights violation and the movement of the tournament from the usual summer season to the winter. An analysis will be performed to identify if any of the previous factors had an impact on the attendance of Qatar 2022 in comparisons to other FIFA World Cup tournments


```python
# importing data

fifa_filepath = "../input/fifa-world-cup-attendance-19302022/FIFA World Cup Attendance.csv"

fifa_data = pd.read_csv(fifa_filepath, index_col='Year')

fifa_data
```

```python
# cleaning up dataset

fifa_data

# removing last row from table
fifa_data.drop(fifa_data.tail(1).index,inplace=True)

fifa_data.tail()
```

## World Cup Analysis: In-Stadium Attendance

The total attendance of Qatar 2022 is in line with the last four World Cup tournments. Interestingly enough, there was an increase of almost 400,000 from the previous tournemnt, Russia 2018, a host nation which was also subjected to various controversies and serves as a good comparison for this analysis. A general overview depicts it as the third most attended tournament in its history, revealing a lack of impact of the aforementioned factors.

```python
# line plot data viz

# setting up figure
plt.figure(figsize=(14,7))
# title
plt.title("FIFA World Cup Attendance")
# line plot showing the fifa world cup attendance
sns.lineplot(data=fifa_data['Total_Attendance'])
# axis labels
plt.ylabel('Total Attendance in Millions')
```

```python
# bar plot data viz

# setting up figure
plt.figure(figsize=(14,7))
# title
plt.title("FIFA World Cup Attendance")
# bar plot showing the fifa world cup attendance
sns.barplot(x=fifa_data.index,y=fifa_data['Total_Attendance'])
# axis labels
plt.ylabel('Total Attendance in Millions')
```

## Update

After extensive research, I've attempted to compile a world cup viewership database, allowing for further analysis into the impact of controversies on the Qatar 2022 World Cup.

```python
# importing data

audience_filepath = "../input/fifa-world-cup-global-audience-reach/FIFA World Cup Global Audience Reach.csv"

audience_data = pd.read_csv(audience_filepath,index_col='Year')

audience_data
```

```python
# cleaning up dataset

# removing last row from table
audience_data.drop(audience_data.tail(1).index,inplace=True)

audience_data.tail()
```

## World Cup Analysis: Global Viewership

There was a 34% increase in the total TV audience viewership for the finals in comparison to the previous year. This followed the positive trend of viewership increase seen after the 2002 South Korea/Japan World Cup tournament. Although the full consolidated report on the Qatar audience data has yet to be pulished, a conclusion could be made of an increase in the total global audience reach. This can be supported from [FIFA's TV audience record breaking claims](https://www.fifa.com/tournaments/mens/worldcup/qatar2022/news/fifa-world-cup-delivering-record-breaking-tv-audience-numbers) of several entertaining group matches such as Spain vs. Germany; which exceeded the audience of any group stage match for the FIFA World Cup in 2018.


```python
# line plot data viz

# setting up figure
plt.figure(figsize=(14,7))
# title
plt.title("FIFA World Cup Global Audience Reach")
# line plot showing the fifa world cup global reach
sns.lineplot(data=audience_data['Global_Reach'])
# axis labels
plt.ylabel('Total In-Home Viewership in Billions')
plt.ylim(0)
```

```python
# bar plot data viz

# setting up figure
plt.figure(figsize=(14,7))
# title
plt.title("FIFA World Cup Finals Viewership")
# bar plot showing the fifa world cup attendance
sns.barplot(x=audience_data.index,y=audience_data['Finals_Reach'])
# axis labels
plt.ylabel('Total Finals Viewership in Billions')
```
